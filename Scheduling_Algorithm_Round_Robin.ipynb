**Algoritmo de Planificación Round Robin (RR)**

```python
# Simulación de algoritmo de planificación de procesos RR (Round Robin)
print('Datos de entrada para Simulación de algoritmo de planificación ')
print('de procesos RR (Round Robin)')
print(' ')

class process(object):
    def __init__(self, id, burst, arrival):
        self.id = id
        self.burst = burst  #Rafaga
        self.arrival = arrival  #Llegada
        self.burst_tmp = burst
        self.wait = 0
        self.return_ = 0
        self.ending = 0

try:
    amount_processes = int(input('Ingrese la cantidad de procesos: '))
    if (amount_processes > 0):
        process_list = []
        for amount_p in range(amount_processes):
            arrival_tmp = -1
            print(f'Proceso número: {amount_p + 1}')
            while (arrival_tmp < 0):
                arrival_tmp = int(
                    input(f'Ingrese el tiempo de llegada del proceso número {amount_p + 1}: '))

            burst_tmp = 0
            while (burst_tmp < 1):
                burst_tmp = int(
                    input(f'Ingrese la carga del proceso número {amount_p+1}: '))
            #id, burst, arrival
            process_list.append(
                process((amount_p + 1), burst_tmp, arrival_tmp))
            print(f'Fin del proceso')
            print()
            print()
        quantum = 0
        while (quantum < 1):
            quantum = int(input('Ingrese el Quantum: '))#Unidad de tiempo de CPU
        quantum_tmp = quantum

#Función para calcular el orden de los procesos segun el tiempo de llegada
        def sortedbyArrivalTime(list):
            for proc in range(1, len(list)):
                item = proc
                while (item > 0 and list[item].arrival < list[item-1].arrival):
                    list[item], list[item-1] = list[item-1], list[item]
                    item = item-1
            return list
        process_list = sortedbyArrivalTime(process_list)  #Se ordena por tiempo de llegada
        mirror_process = len(process_list)
        time = 0
        queue_processes = []
        currently_running_process = None  # proceso actualmente en ejecución
        next_process = 0
        print()
        print()
        print('INICIO DE LA EJECUCIÓN DEL ALGORITMO RR')
        print()
        control = True
        while (mirror_process > 0):
            print(
                f'Tiempo: {time}')
            if (len(process_list) > next_process and time >= process_list[next_process].arrival):
                print(
                    f'El proceso {process_list[next_process].id} se ingreso a la cola de listos')
                queue_processes.append(process_list[next_process])
                next_process += 1
            else:
                if next_process > 0 or len(queue_processes) > 0:
                    if (currently_running_process == None):
                        currently_running_process = queue_processes.pop(
                            0)
                        control = True
                        print(
                            f'Se saca el proceso {currently_running_process.id} de la cola y se ejecuta.')

                    if (control):
                        if (currently_running_process.burst_tmp >= quantum):
                            currently_running_process.burst_tmp = currently_running_process.burst_tmp - quantum
                            print(
                                f'Se resta {quantum} a la rafaga del proceso {currently_running_process.id}')
                            time = time + quantum
                            print(f'Se aumenta {quantum} al tiempo')
                        else:
                            time = time + currently_running_process.burst_tmp
                            print(
                                f'Se aumenta {currently_running_process.burst_tmp} al tiempo')
                            print(
                                f'Se resta {currently_running_process.burst_tmp} a la rafaga del proceso {currently_running_process.id}')
                            currently_running_process.burst_tmp = 0

                        if (currently_running_process.burst_tmp < 1):
                            print(
                                f'Tiempo: {time}')
                            print(
                                f'Proceso {currently_running_process.id} terminado.')
                            currently_running_process.ending = time
                            currently_running_process.return_ = currently_running_process.ending - \
                                currently_running_process.arrival
                            currently_running_process.wait = currently_running_process.return_ - \
                                currently_running_process.burst
                            mirror_process = mirror_process - 1
                            currently_running_process = None

                        else:
                            control = False
                    else:
                        queue_processes.append(currently_running_process)
                        print(
                            f'Se agrega el proceso {currently_running_process.id} que estaba en ejecución a la cola de listos')
                        currently_running_process = None
                else:
                    time += 1
        print()            
        print('Algoritmo finalizado')
        print()
        print()
        print('RESULTADO DE LA SIMULACIÓN')
        total_return = 0
        total_wait = 0
        for process in process_list:
            print(
                f'Proceso #{process.id}, terminado: {process.ending}, tiempo de espera: {process.wait}, retorno: {process.return_}')
            total_return = total_return + process.return_
            total_wait = total_wait + process.wait
            print(f'Promedio de retorno: {total_return/len(process_list)}')
            print(f'Promedio de espera: {total_wait/len(process_list)}')
            print()
    else:
        print('El número de procesos debe ser mayor a 0 (cero) para poder probar el algoritmo.')
except Exception as e:
    print(e)
```
