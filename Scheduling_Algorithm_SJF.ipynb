**Algoritmo de Planificación Shortes Job First (SJF) without arrival time**

```python
# Simulación de algoritmo de planificación de procesos SJF (Shortest Job First)
print('Datos de entrada para Simulación de algoritmo de planificación ')
print('de procesos SJF (Shortest Job First)')
print(' ')

n=int(input('Ingrese la cantidad de procesos: '))
print('Ingrese el ID del proceso y la carga del proceso respectivamente: ')
print('ID CPUtime')

l=[] # lista de la cadena de entrada, dividida, asignada int
for _ in range(n):
  l.append(list(map(int,input().split())))
  
  #ordenar la lista según el burst time en orden ascendente
for i in range(n):
  for j in range(i+1,n):
    if l[i][1]>l[j][1]: 
      l[i],l[j]=l[j],l[i] #intercambio

ct=0 #tiempo de finalización
for i in range(n):
  ct=l[i][1]
  l[i].append(ct) #agregando ct al proceso

print(' ')
print('ID CPUtime CT/TAT WaitingTime')
tat=0  #Tiempo de respuesta del proceso
wt=0 #Tiempo de espera del proceso
for i in range(n):
  print (l[i][0],'   ',l[i][1],'     ',l[i][2],'     ',l[i][2]-l[i][1])
  tat=l[i][2]
  wt=l[i][2]-l[i][1]
print('Tiempo promedio de rafaga de tiempo: ',tat/n)
print('Tiempo promedo de espera: ',wt/n)


```
