**Algoritmo de Planificación First Come, First Served (FCFS)**

```python
# Simulación de algoritmo de planificación de procesos FCFS (First Come First Served)

print('Datos de entrada para Simulación de algoritmo de planificación ')
print('de procesos FCFS (First Come First Served)')
print(' ')
print('ANTES DE INICIAR!! ') 
print('No repita el tiempo de llegada entre cada proceso')
print(' ')

from operator import itemgetter
try:
    amount_processes = int(input('Ingrese el número de procesos: '))
    if(amount_processes > 0):
        processes = {}
        for pr in range(amount_processes):
            print(f'Proceso número {pr + 1}')
            arrival_time = int(input(f'Ingrese el tiempo de llegada del proceso {pr+1}: '))
            burst= int(input(f'Ingrese el valor de carga del proceso {pr+1}: '))
            processes[arrival_time] = burst
            print()
            print()
       
        print('Ordenamos la cola de los procesos según su tiempo de llegada')
        print() #ordenar el diccionario a partir de su arrival_time
        processes_sorted = dict(sorted(processes.items(), key=itemgetter(0)))

        response_time = 0 #tiempo de respuesta
        totalRT = []
        for arrival_time, burst in processes_sorted.items():
            response_time = burst
            print(f'Proceso con tiempo de llegada número: {arrival_time}')
            print(f'Tiempo de respuesta número: {response_time}')
            print('Proceso Finalizado')
            print(' ')
            totalRT.append(response_time)

        contador = 0
        waiting_time = []
        for wt in totalRT:
            contador += 1
            val = wt - contador
            waiting_time.append(val)

        waiting_time = waiting_time[:-1]
        waiting_time.append(0)
        waiting_time.sort()

        sumWT = sum(waiting_time)
        sumRT = sum(totalRT)
        
        averageWT  =  sumWT / len(processes_sorted)
        averageRT = sumRT / len(processes_sorted)
        print(f'El tiempo de espera {waiting_time}, cuenta con un tiempo total de espera de: {sumWT}')
        print(f'El tiempo de espera {waiting_time}, cuenta con un promedio de espera de: {averageWT}')
        print(f'El tiempo de respuesta {totalRT}, cuenta con un promedio de respuesta de: {averageRT}')


    else:
       print('El número de procesos debe ser mayor a 0 (cero) para poder probar el algoritmo.')
except Exception as e:
    print(e)
```
